```python
import numpy as np
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski
```

```python
MCHR1_no_smiles = MCHR1a.drop(columns='Smiles')
```

```python
smiles = []

for i in MCHR1a['Smiles'].tolist():
  cpd = str(i).split('.')
  cpd_longest = max(cpd, key = len)
  smiles.append(cpd_longest)

smiles = pd.Series(smiles, name = 'smiles')
```

```python
MCHR1_clean_smiles = pd.concat([MCHR1_no_smiles,smiles], axis=1)
MCHR1_clean_smiles
```

```python
def lipinski(smiles, verbose=False):

    moldata= []
    for elem in smiles:
        mol=Chem.MolFromSmiles(elem) 
        moldata.append(mol)
       
    baseData= np.arange(1,1)
    i=0  
    for mol in moldata:        
       
        desc_MolWt = Descriptors.MolWt(mol)
        desc_MolLogP = Descriptors.MolLogP(mol)
        desc_NumHDonors = Lipinski.NumHDonors(mol)
        desc_NumHAcceptors = Lipinski.NumHAcceptors(mol)
           
        row = np.array([desc_MolWt,
                        desc_MolLogP,
                        desc_NumHDonors,
                        desc_NumHAcceptors])   
    
        if(i==0):
            baseData=row
        else:
            baseData=np.vstack([baseData, row])
        i=i+1      
    
    columnNames=["MW","LogP","NumHDonors","NumHAcceptors"]   
    descriptors = pd.DataFrame(data=baseData,columns=columnNames)
    
    return descriptors
```

```python
MCHR1_lipinski = lipinski(MCHR1_clean_smiles.smiles)
MCHR1_lipinski.head()
```

```python
MCHR1_combined = pd.concat([MCHR1_clean_smiles,MCHR1_lipinski], axis=1)
MCHR1_combined.head()
```

```python
def norm_value(input):
    norm = []

    for i in input['Standard Value']:
        if i > 100000000:
          i = 100000000
        norm.append(i)

    input['standard_value_norm'] = norm
    x = input.drop('Standard Value', 1)
        
    return x
```

```python
MCHR1_norm = norm_value(MCHR1_combined)
MCHR1_norm
```

```python
def pIC50(input):
    pIC50 = []

    for i in input['standard_value_norm']:
        molar = i*(10**-9) # Converts nM to M
        pIC50.append(-np.log10(molar))

    input['pIC50'] = pIC50
    x = input.drop('standard_value_norm', 1)
        
    return x
```

```python
MCHR1v = pIC50(MCHR1_norm)
MCHR1v.head()
```

```python
MCHR1f = MCHR1v[MCHR1v['class'] != 'intermediate']
MCHR1f.head()
```
